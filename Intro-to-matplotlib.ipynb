## Matplotlib Workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Prepare data
x = [1, 3, 4, 5, 6]
y = [33, 55, 77, 88, 100]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 5))

# 3. Plot data
ax.plot(x, y);

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
    
# 5. Save and show your whole figure
fig.savefig("sample-plot.png")
```

# Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(100), np.random.random(100));
ax3.bar(dic.keys(), dic.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
#Plot to each index
ax[0, 0].plot(x, x**6);
ax[0, 1].scatter(np.random.random(1000), np.random.random(1000));
ax[1, 0].bar(dic.keys(), dic.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should I use? (myplot vs matplotlib OO method)
* For advanced plots use advanced method i.e OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head(10)
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO Method with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
```

```python
# Object Oriented method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Adda a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line in matploz̄lib plot
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Subplot of Age, chol, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(
       xlabel="Age",
       ylabel="Cholesterol")

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add meanline to ax0
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max heart rate")

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add meanline to ax1
ax1.axvline(x=over_50["age"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
